```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


img_path = 'C:/Users/moree/Downloads/img.jpeg'
img = cv2.imread(img_path)


gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


_, im1 = cv2.threshold(gray_img, 50, 255, cv2.THRESH_BINARY)
_, im2 = cv2.threshold(gray_img, 200, 255, cv2.THRESH_BINARY_INV)
_, im3 = cv2.threshold(gray_img, 127, 255, cv2.THRESH_TRUNC)
_, im4 = cv2.threshold(gray_img, 127, 255, cv2.THRESH_TOZERO)
_, im5 = cv2.threshold(gray_img, 127, 255, cv2.THRESH_TOZERO_INV)


plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(2, 3, 2), plt.imshow(im1, cmap='gray'), plt.title('Threshold 1')
plt.subplot(2, 3, 3), plt.imshow(im2, cmap='gray'), plt.title('Threshold 2')
plt.subplot(2, 3, 4), plt.imshow(im3, cmap='gray'), plt.title('Threshold 3')
plt.subplot(2, 3, 5), plt.imshow(im4, cmap='gray'), plt.title('Threshold 4')
plt.subplot(2, 3, 6), plt.imshow(im5, cmap='gray'), plt.title('Threshold 5')

plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

img_path = 'C:/Users/moree/Downloads/img.jpeg'
img = cv2.imread(img_path, cv2.IMREAD_COLOR)

if img is None:
    print("Error: Image not found.")
else:
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Global thresholding
    _, im1 = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)

    # Adaptive thresholding - Mean
    im2 = cv2.adaptiveThreshold(
        gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

    # Adaptive thresholding - Gaussian
    im3 = cv2.adaptiveThreshold(
        gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

    # Plot the images using matplotlib
    plt.figure(figsize=(10, 6))

    plt.subplot(221)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')

    plt.subplot(222)
    plt.imshow(gray_img, cmap='gray')
    plt.title('Gray Image')

    plt.subplot(223)
    plt.imshow(im1, cmap='gray')
    plt.title('THRESH_BINARY')

    plt.subplot(224)
    plt.imshow(im2, cmap='gray')
    plt.title('ADAPTIVE_THRESH_MEAN')

    plt.show()
```
